```python
!pip install easyocr
# 1️⃣ Upload image in Colab
from google.colab import files
uploaded = files.upload()  # choose your image file
img_path = list(uploaded.keys())[0]  # get uploaded file name

# 2️⃣ Import libraries
import easyocr
import cv2
import matplotlib.pyplot as plt

# 3️⃣ Load the image
img = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 4️⃣ Create EasyOCR Reader (English)
reader = easyocr.Reader(['en'])

# 5️⃣ Perform OCR
results = reader.readtext(img_rgb)

# 6️⃣ Display results and draw bounding boxes
for (bbox, text, prob) in results:
    print(f"Detected text: {text} (Confidence: {prob:.2f})")

    # Draw rectangle
    (top_left, top_right, bottom_right, bottom_left) = bbox
    top_left = tuple(map(int, top_left))
    bottom_right = tuple(map(int, bottom_right))
    cv2.rectangle(img_rgb, top_left, bottom_right, (0, 255, 0), 2)

# 7️⃣ Show image with detected text
plt.figure(figsize=(10,10))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
